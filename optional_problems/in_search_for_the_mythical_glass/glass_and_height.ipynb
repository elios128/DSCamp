Before jumping to 3 glass problem, let's see what we can do with unlimited glasses, what would be the best strategy with exceeding resources.

On problems like this, where all info we have is to go higher or lower the best strategy is binary search, every attempt we cann throw out 50% of floors, we cant go higher than that, and on 2 glass 100 floor, binary search gives us results like:

```python
def most_attempts_we_need(n, m, s):
    t = round(n//(2**(m-2)))
    k = 0
    while t >= s:
        t -= s
        s = s - 1 if s> 1 else s
        k+=1
    r = max(t, s-1) #we always choose the worst scenario,  and if remainder is 3 in worst scenario it will break right on that floor and we need to go 1 iteration down and check everything bottom to top

    cost = max(round(n**(1/2)) + (m-2) , k + r + (m-2) )
    return cost, k
```

a new idea, m(n, k) - formula to find how much floors we can check with n glasses and k tries, we need to look for the smallest possible integer k so it gives m >= treshold (num of floors), and we will change the formula, we will look for all possible ways of glass breaking in k steps, and sum them to get the number of floors we can search.

```python
from math import factorial
```

```python
def all_posible_ways_to_break(attempts, glasses):
    result = factorial(attempts)/(factorial(glasses)*factorial(attempts-glasses))
    return result
```

```python
def num_of_floors_we_can_check(attempts, glasses):
    res = 0
    for i in range(1, glasses+1):
        ways_to_break = all_posible_ways_to_break(attempts, i)
        res += ways_to_break
    return res
```

```python
pairs_to_check = [(10, 2), (100, 2), (100,3), (1000, 3), (1000, 5)]

for floors, glasses in pairs_to_check:
    floors_we_can_check = 0
    attempts = glasses #so we dont get negatives in factorials
    while floors_we_can_check < floors:
        floors_we_can_check = num_of_floors_we_can_check(attempts, glasses)
        attempts += 1
    print(f"Floors = {floors}, glasses = {glasses}, least attempts  = {attempts-1}")
```
